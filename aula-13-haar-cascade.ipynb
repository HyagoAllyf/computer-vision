# Algoritmo HAAR CASCADE

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Carraga arquivos de caractersticas
cascadeFace = cv2.CascadeClassifier('C:/Users/hyago/documentos/Meusprojetos/python/computacao_visual/computer-vision/haarcascades/haarcascade_frontalface_default.xml')

imagemOriginal = cv2.imread('C:/Users/hyago/documentos/Meusprojetos/python/computacao_visual/computer-vision/selecao.png')

imagem = cv2.cvtColor(imagemOriginal, cv2.COLOR_BGR2GRAY)

faces = cascadeFace.detectMultiScale(imagem, scaleFactor=1.3, minNeighbors=5, minSize=(30, 30))

# Desenha um retngulo nas faces detectadas

for (x,y,w,h) in faces:
    cv2.rectangle(imagemOriginal, (x, y), (x+w, y+h), (000,255,0), 2)

# Exibir o total de faces detectadas

plt.figure()
plt.imshow(cv2.cvtColor(imagemOriginal, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
